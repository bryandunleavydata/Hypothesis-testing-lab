# <b> Define the null hypothesis (H₀) and alternative hypothesis (H₁):



Null hypothesis (H0): The group's average systolic blood pressure is equal to the regular population's average systolic blood pressure (μ = 120 mm Hg).

Alternative hypothesis (H1): The group's average systolic blood pressure is significantly different from the regular population's average systolic blood pressure (μ ≠ 120 mm Hg)

# <b> Select the significance level (α) for the test. 
How strong the sample evidence must be to conclude that the null hypothesis can be rejected.

```python
Standard Error or SEM = SD / √n
Standard Error = 21.21 mm Hg / √100
Standard Error = 2.121 mm Hg
The SEM (Standard Error of the Mean) is a measure of the variability of sample means around the population mean. 
It quantifies the precision or accuracy of the sample mean as an estimate of the population mean.
```

```python
The margin of error is obtained by multiplying the critical value by the standard error.
Margin of error = 1.984 x 2.121 = 4.208064
```

```python
Degrees of Freedom (DF) = n - 1 = 100 - 1 = 99
```

The mean of a sample is 130.1, SEM 2.121, sample size 100. What is the 95% confidence interval of the mean?

Degrees of freedom (DF) is n−1 = 99, t-value in column for area 0.95 is 1.984 (Number of SD away from the mean). The 95% CI is:

mean − t SEM(upper bounds of confidence interval) to mean + t SEM (lower bounds of cofidence interval)
130.1 − 1.984 x 2.121	to	130.1 + 1.984 x 2.121
125.89	to	134.31

```python
import numpy as np
import matplotlib.pyplot as plt


# Calculate sample mean and standard error
mean = 130.1
std_error = 2.121

# Determine critical value (for a 95% confidence interval)
critical_value = 1.984

# Calculate margin of error
margin_of_error = critical_value * std_error

# Calculate upper and lower bounds
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Plotting
plt.bar(0, mean, yerr=margin_of_error, align='center', alpha=0.5, ecolor='black', capsize=10)
plt.xticks([])
plt.ylabel('Mean')
plt.title('95% Confidence Interval')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the normal distribution
mean = 130.1
std = 21.21
sample_size = 100

data = np.random.normal(mean, std, sample_size)

# Plot the histogram
plt.hist(data, bins=5, density=True, alpha=0.5, color='blue')

# Plot the normal distribution curve
x = np.linspace(mean - 3*std, mean + 3*std, 100)
y = (1/(std * np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mean)/std)**2)
plt.plot(x, y, color='red')

# Set labels and title
plt.xlabel('Systolic Blood Pressure (mm Hg)')
plt.ylabel('Density')
plt.title('Distribution of Systolic Blood Pressure')

# Show the plot
plt.show()
```

```python
Test statistic = t. 
t = (x̄ - μ) / (s / √n)
where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.
t = (130.1 - 120) / (21.21 / √100) = 10.1 / 2.121 = 4.76190 
```

```python
import math

sample_mean = 130.1
population_mean = 120
sample_std_dev = 21.21
sample_size = 100

test_statistic = (sample_mean - population_mean) / (sample_std_dev / math.sqrt(sample_size))
print("Test Statistic:", test_statistic)
```
